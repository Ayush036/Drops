```python
dataset["Location"]=dataset["Location"].map(location_dict)
```

```python
#replace Na in numerical columns with mean for columns with Na ratio higher than 3%:
dataset['WindGustSpeed'].fillna(np.mean(dataset['WindGustSpeed'].dropna().values), inplace = True)
dataset['Pressure9am'].fillna(np.mean(dataset['Pressure9am'].dropna().values), inplace = True)
dataset['Pressure3pm'].fillna(np.mean(dataset['Pressure3pm'].dropna().values), inplace = True)
```

```python
dataset['WindGustDir']= dataset['WindGustDir'].fillna('Unknown')
dataset['WindDir9am']= dataset['WindDir9am'].fillna('Unknown')
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.isnull().sum()
```

```python
# encode categorical values
categorical = ['WindGustDir','WindDir9am','WindDir3pm']
dataset = pd.get_dummies(dataset,columns = categorical,drop_first=True)
```

```python
dataset
```

```python
dataset["Rainfall"].describe()
```

```python
dataset["RainToday"].replace({"Yes":1,"No":0},inplace=True)
```

```python
dataset=dataset.reset_index(drop=True)
```

```python
scale_feat=dataset[['MinTemp','MaxTemp','Rainfall','WindGustSpeed','WindSpeed9am','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am','Pressure3pm','Temp9am','Temp3pm']]
```
