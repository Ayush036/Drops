```python
x=dataset.drop("rain_yes",axis=1)
```

```python
y=dataset.loc[:,"rain_yes"].values
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(x,y)
print(model.feature_importances_) 
```

```python
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(15).plot(kind='barh')
plt.show()
```

```python
d=feat_importances.nlargest(15).index
```

```python
top_features=feat_importances.nlargest(15).index
```

```python
x=x[top_features].values
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
```

```python
def result(algo):
    var=algo
    var.fit(x_train,y_train)
    y_pred=var.predict(x_test)
    return accuracy_score(y_pred,y_test) , f1_score(y_pred,y_test) ,confusion_matrix(y_pred,y_test)




    
    
```

```python
result(KNeighborsClassifier())
```

```python
result(LogisticRegression())
```

```python
result(RandomForestClassifier())
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
import pickle
# open a file, where you ant to store the data
filename = 'rain_model2.sav'
pickle.dump(rf, open(filename, 'wb'))
```

```python
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(x_test, y_test)
print(result)
```
